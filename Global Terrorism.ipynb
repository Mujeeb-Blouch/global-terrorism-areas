```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
encodings_to_try = ['utf-8', 'latin1', 'ISO-8859-1', 'utf-16']
for encoding in encodings_to_try:
    try:
        df = pd.read_csv("Global Terrorism.csv", encoding=encoding)
        print(f"Successfully read the CSV file using {encoding} encoding.")
        break
    except UnicodeDecodeError:
        print(f"Failed to read using {encoding} encoding. Trying the next one.")
```

### DATA VISUALIZATION

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['iyear'], bins=30, kde=True)
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.title('Distribution of Incidents by Year')
plt.show()
```

#### Catagorical Values

```python
group_counts = df['gname'].value_counts()
print(group_counts.head(10))
plt.figure(figsize=(12, 6))
sns.barplot(x=group_counts.head(10).index, y=group_counts.head(10).values)
plt.xticks(rotation=90)
plt.xlabel('Terrorist Group')
plt.ylabel('Number of Incidents')
plt.title('Top 10 Terrorist Groups with the Most Incidents')
plt.show()
```

#### Correlation Analysis 

```python
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

####  Terrorist Attacks Over Time: 

```python
plt.figure(figsize=(12, 6))
attacks_per_year = df['iyear'].value_counts().sort_index()
sns.lineplot(x=attacks_per_year.index, y=attacks_per_year.values)
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.title('Terrorist Attacks Over Time')
plt.show()
```

#### Attack Types 

```python
# Count the number of attacks for each attack type
attack_type_counts = df['attacktype1_txt'].value_counts()

# Plot the distribution of attack types
plt.figure(figsize=(12, 6))
sns.barplot(x=attack_type_counts.index, y=attack_type_counts.values)
plt.xticks(rotation=90)
plt.xlabel('Attack Type')
plt.ylabel('Number of Attacks')
plt.title('Distribution of Attack Types')
plt.show()
```

#### Targets types

```python
target_type_counts = df['targtype1_txt'].value_counts()
plt.figure(figsize=(12, 6))
sns.barplot(x=target_type_counts.index, y=target_type_counts.values)
plt.xticks(rotation=90)
plt.xlabel('Target Type')
plt.ylabel('Number of Attacks')
plt.title('Distribution of Target Types')
plt.show()
```

#### Geographical Trends

```python
country_counts = df['country_txt'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=country_counts.values, y=country_counts.index, palette='viridis')
plt.xlabel('Number of Attacks')
plt.ylabel('Country')
plt.title('Top 10 Countries Most Affected by Terrorism')
plt.show()
```

#### Victim Analysis 

```python
total_killed = df.groupby('iyear')['nkill'].sum()
total_wounded = df.groupby('iyear')['nwound'].sum()
plt.figure(figsize=(12, 6))
plt.plot(total_killed.index, total_killed.values, label='Total Killed')
plt.plot(total_wounded.index, total_wounded.values, label='Total Wounded')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.title('Total Killed and Wounded in Terrorist Attacks (per Year)')
plt.legend()
plt.show()
```

#### Count the number of Killed per years

```python
total_casualties = df.groupby('iyear')[['nkill', 'nwound']].sum()
total_casualties['total_casualties'] = total_casualties['nkill'] + total_casualties['nwound']
most_affected_years = total_casualties['total_casualties'].nlargest(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=most_affected_years.index, y=most_affected_years.values, palette='viridis')
plt.xlabel('Year')
plt.ylabel('Total Casualties (Killed + Wounded)')
plt.title('Top 10 Most Affected Years by Total Casualties')
plt.show()
```

#### Success rate of Attacks 

```python
# Calculate the success rate of attacks
success_rate = df['success'].value_counts(normalize=True) * 100

# Plot the success rate
plt.figure(figsize=(6, 6))
plt.pie(success_rate, labels=['Failed', 'Successful'], autopct='%1.1f%%', colors=['#ff9999','#66b3ff'])
plt.title('Success Rate of Terrorist Attacks')
plt.show()
```

#### Suicide Attacks 

```python
suicide_attack_percentage = (df['suicide'].sum() / len(df)) * 100

print('Percentage of Suicide Attacks: {:.2f}%'.format(suicide_attack_percentage))
```

#### Hostages 

```python
hostages_count = df['ishostkid'].sum()
ransom_count = df['ransom'].sum()
print('Number of Incidents with Hostages: {}'.format(hostages_count))
print('Number of Incidents with Ransom: {}'.format(ransom_count))
```

####  Number of attacks in each country

```python
country_counts = df['country_txt'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=country_counts.values, y=country_counts.index, palette='viridis')
plt.xlabel('Number of Attacks')
plt.ylabel('Country')
plt.title('Top 10 Countries Most Affected by Terrorism')
plt.show()
```

#### Number of Attack in Each Regions

```python
region_counts = df['region_txt'].value_counts()
# Plot the distribution of attacks by region
plt.figure(figsize=(12, 6))
sns.barplot(x=region_counts.values, y=region_counts.index, palette='viridis')
plt.xlabel('Number of Attacks')
plt.ylabel('Region')
plt.title('Distribution of Terrorist Attacks by Region')
plt.show()
```
